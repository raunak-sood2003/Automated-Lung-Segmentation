```python
slices_list = [[148, 212], [160, 242],[214, 318],[183, 262],[170, 260],[188, 244],
               [217, 301],[166, 264],[246, 339],[174, 255],[177, 251],[153, 239],
               [230, 308],[168, 235],[214, 293],[176, 258],[254, 303],[132, 222],[174, 249],[171, 244]]
```

```python
lung_dirs = 'D:/DATA/VESSEL_DATA/lung_CT_numpy'
mask_dirs = 'D:/DATA/VESSEL_DATA/lung_masks_numpy'
```

```python
lower_lung_volume = []
middle_lung_volume = []
upper_lung_volume = []
for idx, pairs in enumerate(slices_list):
    path_lung = os.path.join(lung_dirs, os.listdir(lung_dirs)[idx])
    path_mask = os.path.join(mask_dirs, os.listdir(mask_dirs)[idx])
    lung_npy = np.load(path_lung)
    mask_npy = np.load(path_mask)
    lower_lung_volume.append([lung_npy[:pairs[0], :, :], mask_npy[:pairs[0], :, :]])
    middle_lung_volume.append([lung_npy[pairs[0]:pairs[1], :, :], mask_npy[pairs[0]:pairs[1], :, :]])
    upper_lung_volume.append([lung_npy[pairs[1]:, :, :], mask_npy[pairs[1]:, :, :]])
```

```python
# [(n, 512, 512), (n, 512, 512)] --> [[(512, 512), (512, 512)]...]
lower_lung = []
for paired_volume in lower_lung_volume:
    lung_volume = paired_volume[0]
    mask_volume = paired_volume[1]
    paired_imgs = [[lung_volume[i, :, :], mask_volume[i, :, :]] for i in range(lung_volume.shape[0])]
    for pair in paired_imgs:
        lower_lung.append(pair)
```

```python
upper_lung = []
for paired_volume in upper_lung_volume:
    lung_volume = paired_volume[0]
    mask_volume = paired_volume[1]
    paired_imgs = [[lung_volume[i, :, :], mask_volume[i, :, :]] for i in range(mask_volume.shape[0]-1)]
    for pair in paired_imgs:
        upper_lung.append(pair)
```
